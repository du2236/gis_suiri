### ⑥高潮浸水想定区域図の作成

#### 河川流量

> 一般河川や，背後に人口・資産が集積し，洪水時に相当な流量が想定される河川等においては，高潮時に洪水が同時に発生することを想定し，河川の流量を設定することを基本とする．

> 河川の流量は，河川整備基本方針で定める基本高水を基本とし，現況施設を考慮して設定する．

#### 潮位

> 基準となる潮位（天文潮）は，朔望平均満潮位を基本とする．

> 過去に当該海岸で異常潮位が生じた場合は，異常潮位の各々の最大偏差の平均値を朔望満潮位に加えることとする．


#### 計画高
- 計画高水流量
> 河道を建設する場合に基本となる流量で，基本高水を河道と各種洪水調節施設に合理的に配分した結果として求められる河道を流れる流量
 
 - 計画高水位（H.W.L）
> 計画高水流量が計画断面を流下するときの水位や河道の縦断・横断を考慮して定められた水位で，河道の設計をする基本となる水位．

## 最大規模の高潮の設定

出典：「高潮浸水想定区域図作成の手引き　Ver.2.00 R2.6」

> 想定する<u>台風の中心気圧は，室戸台風</u>を基本とし，既往実績に応じ対象とする海岸における井戸を考慮して台風の中心気圧を増減させる．<u>最大旋衝風速半径，移動速度については，伊勢湾台風を基本</u>とする．
> 各海岸で潮位偏差が最大となるよう，<u>当該地域等で大きな潮位偏差を生じた複数の台風の経路を平行移動させ，想定する台風の経路を設定す</u>ることとする．

###### 用語の説明

越波
> 暴風や高潮により打ち寄せる波が堤防や護岸を超えること

令和二年　高潮浸水想定区域図についての説明資料　大阪府　湾岸局　より

高潮
> 台風や発達した低気圧が通過するとき，海水面（潮位）が大きく上昇することがあり，これを「高潮」という．高潮は主に<u>「気圧低下による吸い上げ効果」</u>と<u>「風による吹き寄せ効果」</u>が原因となり起こる．また，満潮と高潮が重なると潮位は一層上昇する．

気圧低下による吸い上げ効果
> 台風や低気圧の中心では気圧が周辺より低いため，気圧の高い周辺の空気は海水を押し下げ，中心付近の空気が海水を吸い上げるように作用する結果，海面が上昇する．気圧が1hPa下がると，潮位は約1cm上昇するといわれる．

風による吹き寄せ効果
> 台風や低気圧に伴う強い風が沖から海岸に向かって吹くと，海水は海岸に吹き寄せられ，海岸付近の海面が上昇する．この効果による潮位の上昇は風速の２乗に比例する．また，遠浅の海や風が吹いてくる方向に開いた湾の場合，地形が海面上昇を助長させるように働き，特に潮位が高くなる．

浸水域
> 高潮や高波，洪水に伴う越波・越流によって海岸や河川からの氾濫水により浸水する範囲

浸水深
> 陸上の各地点で水面がもっとも高い位置にきたときの地盤面から水面までの高さ

高潮偏差
> 天体の動きから算出した「天文潮位」と気象等の影響を受けた実際の潮位との差を潮位偏差といい，特に台風等が原因であるものを「高潮偏差」という．

朔望平均満潮位
> 朔（新月）および望（満月）の日から５日以内に現れる各月の最大満潮面の平均値

異常潮位
> 黒潮の蛇行等様々な理由により潮位偏差が高い（あるいは低い）状態が数週間続く現象

高潮水位
> 「朔望平均満潮位＋異常潮位」に加え，台風等に伴う高潮偏差の高さを表したもので，台風襲来時に想定される海水面の高さを示す

TP(Tokyo Peil)
> 東京湾平均海面といい，全国の標高の基準となる海水面の高さ

高潮浸水想定区域図の背景
> 近年の堤防等の施設設計規模を超える災害の頻発を受けて，人命を守り社会経済への影響を最小限に抑えるため，平成２７年５月に水防法が改正された．高潮浸水想定区域図は，この改正された水防法に基づき作成されるものである．

## 気圧場・風場の推算

### 気圧場の推算モデル

高潮計算や波浪計算では，外力となる台風内の気圧分布と風の分布をいかに表現するかが計算精度に大きく影響する．

手引きに準じて，台風による気圧分布を推算するモデルとして一般的なMyersの式を用いた．
台風域内を同心円状の気圧であると仮定したモデルである．

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        P(r) = P_c + \varDelta P exp \left( - \cfrac{r_0}{r} \right) 
    \end{align}
$
</span>
<br>
<br>

ここで，$P(r)$:台風中心から$r km$離れた位置での気圧　$P_c$:台風中心での気圧（hPa）

$\varDelta P$:気圧低下量($= P_\infty - P_c$)(hPa)　$P_\infty$:$r = \infty$ における位置での気圧(hPa) = 1010hPa

$r_0$:最大旋衝風速半径であり，台風中心から最大風速発生地点までの距離に相当するパラメータ

風は動いている空気であり，水平方向の空気の流れ，空気分子が一定の方向に一斉に動く現象である．
空気分子の密度が高い領域は気圧が高く，低いところは気圧が低い．
気体のように流動性のある物質は，密度を均一にしようとして，つまり同じ気圧にしようとして空気分子の移動が始まり風が発生する．

気圧傾度
> 気圧差 / 距離差

気圧傾度力
> 気圧の高いところから低いところに向かって働く力．$\cfrac{1}{\rho} \cfrac{\partial P}{\partial r}$

コリオリ力　
> 地球の自転効果を表す見かけ上の力．北半球では物体の運動方向の直角右向きに働き，南半球では逆に左向きに働く．空気も物体のため，空気の運動である風の向きもコリオリ力によって右へ右へ変わっていく．$F = m * 2\Omega Vsin \varPhi$

地衝風平衡
> 気圧傾度力とコリオリ力が釣り合った状態

傾度風
> コリオリ力＋気圧傾度力＋遠心力

地上風
> コリオリ力＋気圧傾度力＋摩擦力

旋衝力
> 気圧傾度力＋遠心力 $\cfrac{mV^2}{r}$




地衝風は等圧線に対して平行に吹く風

低気圧性循環式

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        \cfrac{V^2}{r} + fV = \cfrac{1}{\rho} \cfrac{\partial P}{\partial r}
    \end{align}
$
</span>
<br>
<br>

rをかけてVに関する二次方程式として表すと

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        V^2 + frV - r  \cfrac{1}{\rho} \cfrac{\partial P}{\partial r} = 0
    \end{align}
$
</span>
<br>
<br>
<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        \cfrac{\partial P}{\partial r} = \cfrac{\varDelta P}{r^2} r_0 exp \left( \cfrac{-r_0}{r} \right)
    \end{align}
$
</span>
<br>
<br>

よって傾度風速の分布$U_1(r)$は

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        U_1(r) = -\cfrac{rf}{2} + \sqrt{\left( \cfrac{rf}{2} \right)^2 + \cfrac{\varDelta P}{\rho_a} \cfrac{r_0}{r} exp \left( -\cfrac{r_0}{r} \right)}
    \end{align}
$
</span>
<br>
<br>

ここで，$f = 2 \Omega sin \Phi$ はコリオリ係数であり,$\Phi$は緯度，$\Omega$は角速度($ = 1  /24 * 3600$)
，$\rho_a = 1.22 (kg / m^3)$は空気の密度である．

### 風場の計算モデル

海面上10mの風速($U_{10}$)は傾度風速$U_1(r)$と台風の移動に伴う風速U_2(r)にそれぞれ海面摩擦を考慮した変換係数$C_1,C_2$を乗じ，傾度風の風向きを約30°台風の中心方向に傾け，ベクトル和で求めることとする．

台風の移動に伴う風速$U_2(r)$は，$V_T$を台風の移動速度とすると

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
        U_2(r) = \cfrac{U_1(r)}{U_1(r_0)} V_T
    \end{align}
$
</span>
<br>
<br>

考慮する地点の台風の中心からの距離における傾度風速と最大旋衝風速半径における傾度風速との比に台風の移動速度を乗じることで求める．

また，

<br>
<br>
<span style="font-size:150%;">
$
    \begin{align}
       &V_1 = C_1 U_1(r) \\ 
       &V_2 = C_2 \cfrac{U_1(r)}{U_1(r_0)} V_T \\ 
       &U_{10} = V_1 cos(\beta - 60°) + V_2 cos(90° - \beta + \theta)
    \end{align}
$
</span>
<br>
<br>

これらの変換係数$C_1,C_2$には一般に一定値として，0.6～0.7が用いられており，実績台風の検証によって定める．

### 砕波
> 沖合からの波が岸に近づき水深が浅くなるにつれて波長が縮むと同時に波高が増し，やがて波形が不安定となって前方に崩れ落ちる．砕波にともない平均海面が上昇する現象のことを波浪効果

### ラディエーションストレス
> 波と流れが共存するとき，波の存在により流体内に生じる応力

## 波が発達する条件

 1. ; 風が強く吹くこと(風速：Wind Speed)　風速が大きいほど発達は大きい．大まかにいえば，波高は風速の2乗に比例する
 2. ; 風が吹き続く時間がながいこと（吹続時間：Duration）ほぼ一様な風が吹いている距離
 3. ; 風が吹き続く距離が長いこと（吹送距離：Fetch）
 
### 風浪（発達過程の波）
> 海上で吹いている風によって生じる波　波が進む速さ＜風速　風からエネルギーを受け発生・発達している波．
 - 個々の波形は不規則で先端がとがっている
 - 周期が短く，変化が激しい

### うねり（減衰過程の波）
> 風浪が風の吹かない領域まで進んだり，海上の風が弱まったり風向きが変化するなどして風による発達がなくなった後に残される波　風浪が減衰しながら伝播する波

 - 周期的で丸みを帯び，正弦波に近い
 - 波高はほぼ一定
 - 短周期の波に比べ，同じ波高でも1波の持つエネルギーが大きい
 - 静穏な天候でも，遠方での後天を要因とするうねりが伝播

### 波浪
> 海洋表面の波動のうち，風によって発生した周期が1-30秒程度のもの．風浪とうねりの総称

### 有義波
> 一点においてある時間に観測したN個の波のうち，波高の高いほうからN/3個までの波について波高・周期を平均したもの．

### 波浪の一生

 1. 風による波（風浪）の発生・発達
 > 風（風速，吹送距離，吹続時間）により発達
 2. 風浪からうねりへの変化
 > 発達した風浪が，粘性等により波高が減少し周期が長めにシフトする
 3. 減衰
 > 逆風や地形による変形・屈折による砕波
 


### 死水域
> 河道内の水面部分で流れのない場所あるいは流れがあっても渦状の場所で，流量の疎通に関係のない場所の部分．主に河道の急拡，急縮，湾曲，構造物の陰，樹木の密生等により生じる．一般に，同一河川であっても流量規模によって死水域が異なる．

#### 湾処

#### 灘
